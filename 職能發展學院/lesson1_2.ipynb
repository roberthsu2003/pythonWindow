```python
import requests
url = "https://tcgbusfs.blob.core.windows.net/dotapp/youbike/v2/youbike_immediate.json"
response = requests.get(url)
dataList = response.json()
```

```python
for item in dataList:
    print(f"站名:{item['sna']}")
    print(f"可借:{item['sbi']}")
    print(f"可還:{item['bemp']}")
    print(f"更新時間:{item['mday']}")
    print("=======================================")
```

```python
outterList = []
for item in dataList:
    innerlist = [item['sna'],item['sbi'],item['bemp'],item['mday']]
    outterList.append(innerlist)

print(outterList)
    
```

```python
import csv
with open('output.csv', 'w', newline='',encoding="utf-8") as file:
     # 建立 CSV 檔寫入器
    writer = csv.writer(file)
    writer.writerow(['站名', '可借', '可還', '更新時間'])
    for item in dataList:
        writer.writerow([item['sna'],item['sbi'],item['bemp'],item['mday']])
    
print("存檔成功")
```
